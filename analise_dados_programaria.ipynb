#Importações

```python
import pandas as pd
```

```python
dados = pd.read_excel('/content/drive/MyDrive/Programaria/planilha_modulo3.xlsx')
df = pd.read_excel('/content/drive/MyDrive/Programaria/licitacoes.xlsx')
```

#Funções

```python
dados.head()#mostra primeiros dados
dados.tail()#mostra ultimos dados
dados.columns#mostra quantidade e nome de cada coluna
dados.describe#mostra descrição da tabela
dados.shape#mostra quantidade de linhas e colunas
dados.info#mostra informações da tabela
len(dados)#mostra quantidade de linhas
dados['SALARIO'].nunique()#mostra quantidade de valores únicos
dados['SALARIO'].unique()#mostra valores únicos
dados.groupby('REGIAO ONDE MORA')['ID'].count()#Groupby: agrupa pela coluna escolhida; Count: faz a contagem das linhas
pd.pivot_table(dados, index='GENERO', columns='GESTOR?', values='ID', aggfunc='count')#Cricao de tabela dinamica
```
